## Structure Description:

As a new task appears the model will increase a new neural network refered to that. The new neural network will initilize its weights from the weights of the previous net. Also besides taking its own new task training data, the new net will take 80% of the training data from the previous one, 80% from the two before, 60% from the three before and so on. By mingling the training data, the most updated net will cover the trainning range from the previous. When the model makes prediction, it will use combine the prediction from the most current net and also some of the previous nets. There are several prediction combination methods of the model.

## 1. Addition Approach

### Idea

The idea of this approach is trying to interfere current net's prediction by adding some of the previous net's predictions on detecting previous task categories. ## 3. Frequency Approach

### Idea

For every case, when a task number (<5) net makes the prediction, the model will record its prediction and all the predictions every net one unit before. The model will then pick the most frequently (higher probability) digit prediction as the answer. If the task number of the net is biger than 5, for the sake of computational expense, the model will only record the predictions of every net two units before. This method is supposed to let the model make more precise prediction when the task number is low and also maintain efficiency when the task number is high.

## 4. Detector

### Idea

Based on the experiences and failure of previous model, I want to make a detector so it can tell every net if the testing feature is in its prediction range or not. By doing this, I can successfully prevent the early nets' noise problem, so the early nets will not output predictions about some digits they have never seen before. 

```python
from IPython.display import Image 
Image(filename='taskid_4.png')
```

I find out the hidden layer responses are separate when the net's input is in its prediction range or not. In this example graph, the blue stream is the hidden layer response when the input is in the net's perdition range (exp: 4 for net5). The green stream is the response when the input is outside the net's perdition range (exp: 7 for net5). In the further implemented model, I make a detector for every net so when any net involves in any combination prediction, its detector will tells it if the feature is in its range or not.
